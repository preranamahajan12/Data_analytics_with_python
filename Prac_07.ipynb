```python
import numpy as np
from scipy.stats import norm
```

```python
# 1)
xbar=32
mu0=30
sigma=10
n=30

z=(xbar - mu0)/(sigma/np.sqrt(n))
print("z statistics=",z)
```

```python
#2)
p_value=1-norm.cdf(z)
print("p-value=",z)
```

```python
#3)Decision rule
alpha=0.05
if p_value<alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")
```

```python
import numpy as np
from scipy.stats import chi2
```

```python
#1)Generate random sample
np.random.seed(1)
data=np.random.normal(50,5,20)
```

```python
#2) sample variance
s2=np.var(data,ddof=1)
print("Sample Variance=",s2)
```

```python
#3)Hypothesis
sigma0_sq=25 
alpha=0.05

#4)chi-square statistics
n=len(data)
chi_sq=(n-1)*s2/sigma0_sq
print("Chi-square statistics=",chi_sq)
```

```python
#5)critical values
lower=chi2.ppf(alpha/2,n-1)
upper=chi2.ppf(1-alpha/2,n-1)
print("Critical values=",lower,upper)
```

```python
#6)p_value
p_value=2*min(chi2.cdf(chi_sq,n-1),1-chi2.cdf(chi_sq,n-1))
print("p_value=",p_value)
```

```python
#7)95% CI for variance
ci_lower=(n-1)*s2/chi2.ppf(1-alpha/2,n-1)
ci_upper=(n-1)*s2/chi2.ppf(alpha/2,n-1)
print("95% CI =",(ci_lower,ci_upper))
```

```python
#8) Inference using CI
if sigma0_sq>=ci_lower and sigma0_sq<=ci_upper:
    print("Fail to reject H0")
else:
    print("Reject H0")
```
