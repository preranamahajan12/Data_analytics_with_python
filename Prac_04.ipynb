# 4 Random Sampling Techniques and Estimation

## a) Simple Random Sampling with and without Replacement


```python
import numpy as np
import pandas as pd
# a)
Y=np.array([12,15,18,20,22,25,27,30,32,35,14,16,19,21,23,26,28,31,33,36,13,17,24,29,34,37,38,39,40,42])
N=len(Y)
```

```python
muY=np.mean(Y)
sig2Y=np.var(Y,ddof=0)
print(f"Population Mean:{muY:.4f} and Population Variance: {sig2Y:.4f}")
```

```python
np.random.seed(123)
sample_wr=np.random.choice(Y,8,replace=True)

Ybar_wr=np.mean(sample_wr)
sig2_wr=np.var(sample_wr,ddof=1)
print(f"Sample Mean:{Ybar_wr:.4f} and sample Variance: {sig2_wr:.4f} for with replacement")
```

```python
np.random.seed(123)
sample_wor=np.random.choice(Y,8,replace=False)

Ybar_wor=np.mean(sample_wor)
sig2_wor=np.var(sample_wor,ddof=1)
print(f"Sample Mean:{Ybar_wor:.4f} and sample Variance: {sig2_wor:.4f} for without replacement")
```

```python
# 7)
VYbar_wr=[]
VYbar_wor=[]
for _ in range(1000):
     VYbar_wr.append(np.mean(np.random.choice(Y,8,replace=True)))
     VYbar_wor.append(np.mean(np.random.choice(Y,8,replace=False)))
VYbar_wr
```

```python
pd.DataFrame({"Method":["SRSWR","SRSWOR"],
              "Means":[np.mean(VYbar_wr),np.mean(VYbar_wor)],
              "Var":[np.var(VYbar_wr),np.var(VYbar_wor)],
              "MSE":[np.mean((VYbar_wr-muY)**2),np.mean((VYbar_wor-muY)**2)]})
```

## b) Stratified Random Sampling

```python
import numpy as np
#Starata
s1=np.array([10,12,14,15,16,18,20,22])
s2=np.array([25,27,28,30,32,34,35,36])
s3=np.array([40,42,45,47,48,50,52,55])

Y = np.concatenate([s1,s2,s3])
N = len(Y)

print("Population Mean:",np.mean(Y))


```

```python
n = 6
N1,N2,N3 = len(s1),len(s2),len(s3)

n1=int(n * N1/N)
n2=int(n * N2/N)
n3=n-n1-n2
s1 = np.random.choice(s1,n1,replace=False)
s2 = np.random.choice(s2,n2,replace=False)
s3 = np.random.choice(s3,n3,replace=False)

strat_mean = (N1/N)*np.mean(s1) + (N2/N)*np.mean(s2) + (N3/N)*np.mean(s3)
print("\nStratified Mean (Proportional):",
      strat_mean)
```
